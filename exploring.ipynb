```python
import pandas as pd

leetcode = pd.read_csv(
    "leetcode-problem-dataset/leetcode_dataset.csv"
)

codeforces = pd.read_csv(
    "codeforces-competitive-programming-dataset/raw_dataset.csv"
)

codechef = pd.read_csv(
    "codechef-competitive-programming/questions.csv"
)

def inspect_dataset(name, df):
    print(f"\n{name}")
    print("=" * len(name))

    print(f"Rows: {df.shape[0]}")
    print(f"Columns: {df.shape[1]}")
    print(f"Data example:\n {df.iloc[0]}")

    summary = pd.DataFrame({
        "column": df.columns,
        "dtype": df.dtypes.astype(str).values,
        "missing": df.isna().sum().values,
        "missing_%": (df.isna().mean() * 100).round(2).values,
        "unique": df.nunique().values,
    })

    return summary

leetcode_summary = inspect_dataset("LeetCode", leetcode)
display(leetcode_summary)

codeforces_summary = inspect_dataset("Codeforces", codeforces)
display(codeforces_summary)

codechef_summary = inspect_dataset("CodeChef", codechef)
display(codechef_summary)
```

# Recommended Merged Dataset Schema

The safest merged dataset should be a **union (concat)** dataset rather than a **join**, because **LeetCode**, **Codeforces**, and **CodeChef** do not share a common problem ID.

Use a **normalized schema** where platform-specific columns are mapped into common names.

## Recommended Core Columns

| Merged Column | LeetCode | Codeforces | CodeChef | Why Use It |
|---------------|----------|------------|-----------|------------|
| **platform** | `"leetcode"` | `"codeforces"` | `"codechef"` | Required to know where each row came from. |
| **source_id** | `id` | `problem_id` or `short_id` | `QCode` | Each dataset has a unique identifier. Useful for deduplication and traceability. |
| **title** | `title` | `problem_name` | `Title` | All three have problem names with almost no missing values. |
| **url** | `url` | `problem_link` | `link` | Same semantic role. CodeChef links are relative, so normalize them to `https://www.codechef.com/problems/...`. |
| **statement** | `description` | `problem_statement` | `statement` | Main problem text. Best common feature for NLP and ML modeling. |
| **difficulty_raw** | `difficulty` | `problem_dificulty` | `level` | All three encode difficulty, but their scales differ. Keep the original value. |
| **difficulty_normalized** | *derived* | *derived* | *derived* | Useful for cross-platform comparison (e.g., `easy`, `medium`, `hard`). |
| **tags_raw** | `related_topics` | `problem_tags` | `Tags` | Closest shared topic/category field. Useful for classification and retrieval tasks. |
| **editorial_url** | `solution_link` | `editorial_link` | `Editorial` | Closest shared explanation/editorial link. Missing for many LeetCode rows, but valuable when available. |

```python
import pandas as pd

leetcode = pd.read_csv(
    "leetcode-problem-dataset/leetcode_dataset.csv"
)

codeforces = pd.read_csv(
    "codeforces-competitive-programming-dataset/raw_dataset.csv"
)

codechef = pd.read_csv(
    "codechef-competitive-programming/questions.csv"
)

def show_difficulties(name, arr):
    print(f"\n{name}")
    print("=" * len(name))

    for i in arr:
        print(i)

# Different levels to normalize:

show_difficulties("leetcode", leetcode["difficulty"].unique())

show_difficulties("codeforces", codeforces["problem_dificulty"].unique())

show_difficulties("codechef", codechef["level"].unique())

def normalize_codeforces(diff):
    if diff <= 1200:
        return "Easy"
    elif diff <= 1600:
        return "Medium"
    return "Hard"

def normalize_codechef(diff):
    if diff == "easy" or diff == "beginner":
        return "Easy"
    elif diff == "hard" or diff == "challenge":
        return "Hard"
    return "Medium"

print('\n\n')

# Normalized
codeforces_diff = codeforces["problem_dificulty"].unique()
for i in codeforces_diff:
    print(normalize_codeforces(i))

print('\n')

codechef_diff = codechef["level"].unique()
for i in codechef_diff:
    print(normalize_codechef(i))
```
