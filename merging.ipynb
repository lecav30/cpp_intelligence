```python
import pandas as pd

leetcode = pd.read_csv(
    "leetcode-problem-dataset/leetcode_dataset.csv"
)

codeforces = pd.read_csv(
    "codeforces-competitive-programming-dataset/raw_dataset.csv"
)

codechef = pd.read_csv(
    "codechef-competitive-programming/questions.csv"
)

leetcode_part = pd.DataFrame({
    "platform": "leetcode",
    "source_id": leetcode["id"],
    "title": leetcode["title"],
    "url": leetcode["url"],
    "statement": leetcode["description"],
    "difficulty_raw": leetcode["difficulty"],
    "difficulty_normalized": leetcode["difficulty"].str.lower(),
    "tags_raw": leetcode["related_topics"],
    "editorial_url": "https://www.leetcode.com" + leetcode["solution_link"].fillna(""),
})

def normalize_codeforces(diff):
    if diff <= 1200:
        return "easy"
    elif diff <= 1600:
        return "medium"
    return "hard"

codeforces_diff_normalized = [normalize_codeforces(i) for i in codeforces["problem_dificulty"]]

codeforces_part = pd.DataFrame({
    "platform": "codeforces",
    "source_id": codeforces["short_id"],
    "title": codeforces["problem_name"],
    "url": codeforces["problem_link"],
    "statement": codeforces["problem_statement"],
    "difficulty_raw": codeforces["problem_dificulty"],
    "difficulty_normalized": codeforces_diff_normalized,
    "tags_raw": codeforces["problem_tags"],
    "editorial_url": codeforces["editorial_link"],
})


def normalize_codechef(diff):
    diff_lower = diff.lower()
    if diff_lower == "easy" or diff_lower == "beginner":
        return "easy"
    elif diff_lower == "hard" or diff_lower == "challenge":
        return "hard"
    return "medium"

codechef_diff_normalized = [normalize_codechef(i) for i in codechef["level"]]
    
codechef_part = pd.DataFrame({
    "platform": "codechef",
    "source_id": codechef["QCode"],
    "title": codechef["Title"],
    "url": "https://www.codechef.com" + codechef["link"].fillna(""),
    "statement": codechef["statement"],
    "difficulty_raw": codechef["level"],
    "difficulty_normalized": codechef_diff_normalized,
    "tags_raw": codechef["Tags"],
    "editorial_url": codechef["Editorial"],
})

merged = pd.concat([leetcode_part, codeforces_part, codechef_part], ignore_index=True)

try:
  merged.to_csv("competitive_programming_dataset.csv", index=False)
except Exception as error:
    print(f"Something went wrong: {error}")
else:
  print("competitive_programming_dataset.csv created succesfully!")
```
