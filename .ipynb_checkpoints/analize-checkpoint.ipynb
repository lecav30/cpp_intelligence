```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

merged = pd.read_csv("competitive_programming_dataset.csv")


def print_title(title, lines=3):
    print(f"\n{'-' * lines} {title} {'-' * lines}")
    print(f"{'-' * (len(title) + lines * 2 + 2)} \n")
    
# merged.isna().sum()
# merged.duplicated().sum()
# merged.dtypes
# merged.nunique()

print_title("Main information")
merged.info()
# print(merged.head())
for platform in merged["platform"].unique():
    print(f"\n{platform}")
    print(merged[merged["platform"] == platform].head(1))

# Columns

print_title("Columns")
print(merged.columns)

# Problems by platform

print_title("Problems by platform")
print(merged["platform"].value_counts())

# Null values by column

print_title("Null values by column")
print(merged.isna().sum())

# How many have editorial_url

print_title("How many have editorial_url")
print(merged["editorial_url"].count())

# Distribution difficulty_normalized

print_title("Distribution difficulty_normalized")

print(merged["difficulty_normalized"].value_counts())

diffs_normalized = (
    merged["difficulty_normalized"]
    .value_counts()
    .sort_index()
)

fig, ax = plt.subplots()

ax.bar(
    diffs_normalized.index, # x 
    diffs_normalized.values, # y
    edgecolor="white",
    linewidth=0.7
)

ax.set_xlabel("Difficulty")
ax.set_ylabel("Number of problems")
ax.set_title("Problems by normalized difficulty")

plt.show()
```
