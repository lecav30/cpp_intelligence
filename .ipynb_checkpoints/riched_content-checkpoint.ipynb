```python
import pandas as pd
import numpy as np
import re

merged = pd.read_csv("competitive_programming_dataset.csv")
```

```python
title_length = [len(i) for i in merged["title"].values]

print(title_length[:10])
```

```python
statement_length = [len(i) for i in merged["statement"].values]

print(statement_length[:10])
```

```python
def word_count(statement):
    words = statement.split()
    return len(words)

statement_word_count = merged["statement"].apply(word_count)

print(statement_word_count[:10])
```

```python
def normalize_tags(tags):
    if pd.isna(tags):
        return None

    tags = str(tags).lower()
    tags = re.sub(r"[\[\]'\"]", "", tags)
    tags = re.sub(r"\s*,\s*", ", ", tags)
    tags = tags.strip()

    return len(tags.split(','))

num_tags = merged["tags_raw"].apply(normalize_tags)

print(merged["tags_raw"].head(5))
print(num_tags.head(5))
```

```python
def has_tag(tags, target_tags):
    if pd.isna(tags):
        return 0

    tags = str(tags).lower()
    tags = re.sub(r"[\[\]'\"]", "", tags)

    tag_list = [tag.strip() for tag in tags.split(",")]
    target_tags = [tag.lower().strip() for tag in target_tags]

    return int(any(target in tag_list for target in target_tags))

has_dp = merged["tags_raw"].apply(lambda tags: has_tag(tags, ["dp", "dynamic-prog", "dynamic programming"]))
has_graph = merged["tags_raw"].apply(lambda tags: has_tag(tags, ["graph", "graphs"]))
has_greedy = merged["tags_raw"].apply(lambda tags: has_tag(tags, ["greedy"]))
has_math = merged["tags_raw"].apply(lambda tags: has_tag(tags, ["math", "maths", "mathematics"]))
```

```python
def avg_word_len(statement):
    words = statement.split()
    return sum(len(word) for word in words) / len(words)

avg_word_length = merged["statement"].apply(avg_word_len)

print(avg_word_length[:10])
```

```python
enriched_dataset = merged.copy()

enriched_dataset = enriched_dataset.assign(
    title_length=title_length,
    statement_length=statement_length,
    statement_word_count=statement_word_count,
    num_tags=num_tags,
    avg_word_length=avg_word_length,
    has_dp=has_dp,
    has_graph=has_graph,
    has_greedy=has_greedy,
    has_math=has_math,
)

print(enriched_dataset.head(5))

enriched_dataset.to_csv("enriched_dataset.csv", index=False)
```
